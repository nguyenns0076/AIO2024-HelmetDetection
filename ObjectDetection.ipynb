## **5.Traning**

```python
%cd /content/yolov10
```

```python
from ultralytics.models import YOLOv10

MODEL_PATH = 'yolov10n.pt'
model = YOLOv10(MODEL_PATH)
```

```python
model.info()
```

```python
YAML_PATH = '../safety_helmet_dataset/data.yaml'
EPOCHS = 5
IMG_SIZE = 320
BATCH_SIZE = 8

model.train(data=YAML_PATH,
            epochs=EPOCHS,
            imgsz=IMG_SIZE,
            batch=BATCH_SIZE)
```

echo là 1 vòng train: data -> model (học đc gì đó) -> thay đổi trọng số. = 1 epoch, hay nói là mô hình nhìn thấy data

batch (chia ra bao nhiêu cặp, nhỏ hơn số lượng dữ liệu n, vd dataset = 100, batch_size = 2 => có 50 batch -> 50 kinh nghiệm -> 1 epoch) : cho học theo cặp (n-1, n) -> học đc gì đó -> cập nhật trọng số (w) hay nói cách khác là chia nhỏ ra để học

## **6.Dự đoán - Predict**

```python
from google.colab.patches import cv2_imshow
import cv2

TRAINED_MODEL_PATH = 'runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

IMAGE_URL = '/content/safety_helmet_dataset/test/images/helmet-10-_jpg.rf.eee985865a1a52407316f0c9ea7a9b3e.jpg'
CONF_THRESHOLD = 0.3

results = model(source=IMAGE_URL,
                conf=CONF_THRESHOLD,
                imgsz=IMG_SIZE)

annotated_img = results[0].plot()
cv2_imshow(annotated_img)
```

training = bắt đầu với trọng số ngẫu nhiên và train model trên tasks mới

pre-training = (học từ không có kiến thức) bắt đầu với trọng số ngẫu nhiên và train model với dataset cực lớn - cực nhiều task

fine-tuning = (học từ kiến thức có sẵn) bắt đầu với trọng số của pre-trained và train model trên task mới

out of mem = giảm batch-size

epoch (giống như khi mô hình hội tụ) khi đã hội tụ rồi thì học nhìu hơn (epoch) nhiều hơn thì cũng kh hơn đc => chỉ nên phù hợp thôi

mAP/accuracy = độ đo đánh giá tốt hay không, với bài detection này thì là xài mAP
học rồi thì không đánh giá, phải giải trên dữ liệu chưa học
